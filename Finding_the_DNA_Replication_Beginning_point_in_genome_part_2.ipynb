##   1.1 The Simplest Way to Replicate DNA

Now, let's look at how DNA replication works in more detail. Imagine a circular chromosome with two DNA strands that are like mirror images of each other. These strands are twisted together and run in opposite directions.

**Starting Point (ori):**

The replication process begins at a specific spot called "ori" (origin of replication). Here, the two DNA strands start to unwind.

**Unwinding and Replication Forks:**

As the strands unwind, they create two "replication forks." These forks move outward in both directions around the chromosome.

**Separation:**

The replication process continues until the two strands are completely separated at a point called the "replication terminus" (denoted as ter). This terminus is located roughly on the opposite side of the chromosome from where the replication started (ori).



**Figure:** Four imaginary DNA polymerases at work replicating a chromosome as the replication forks extend from ori to ter. The blue strand is directed clockwise, and the green strand is directed counterclockwise.

An important thing to know about DNA replication is that the copying doesnâ€™t wait for the two original strands to fully separate. Instead, the copying starts while the strands are still coming apart.

There are four DNA polymerases (the enzymes that copy DNA), and each one works on one half of the DNA strands. All four start copying at the origin (called ori) and work their way around the chromosome.

Before a DNA polymerase can start copying, it needs a primer  a very short piece of DNA (shown in red in the figure) that attaches to the original strand and helps the polymerase get started.

As the strands separate, each DNA polymerase adds new nucleotides starting from the primer and moves along the DNA from ori to the end point called ter, going either clockwise or counterclockwise.

When all four DNA polymerases reach ter, the entire chromosome has been copied. This results in two complete pairs of DNA strands, and the cell is then ready to divide.




**Figure:** Replication is complete.

it assumes that DNA polymerases can copy DNA in either direction along a strand of DNA (i.e., both 5' â†’ 3' and 3' â†’ 5'). However, nature has not yet equipped DNA polymerases with this ability, as they are unidirectional, meaning that they can only traverse a template strand of DNA in the 3' â†’ 5' direction. Notice that this is opposite from the 5' â†’ 3' direction of DNA.

<span style="background-color: orange; color: black;">**If you were a unidirectional DNA polymerase, how would you replicate DNA? How many DNA polymerases would be needed to complete this task?**</span>

If I were a unidirectional DNA polymerase (like real ones in nature), I could only move along the template strand in the 3â€² â†’ 5â€² direction, synthesizing the new strand in the 5â€² â†’ 3â€² direction.

So, how would I replicate DNA?

- DNA is double-stranded and antiparallel, meaning one strand runs 5â€² â†’ 3â€² and the other runs 3â€² â†’ 5â€².

- Since I can only read 3â€² â†’ 5â€², I can easily replicate one strand continuously â€” this is called the leading strand.

- But the other strand (lagging strand) runs 5â€² â†’ 3â€², so I cannot replicate it continuously.

- Instead, I'd replicate it in short fragments (called Okazaki fragments) in the 5â€² â†’ 3â€² direction by jumping back as the fork opens up more DNA.

The unidirectionality of DNA polymerase requires a major revision to our naive model of replication. Imagine that you decided to walk along DNA from ori to ter. There are four different half-strands of parent DNA connecting ori to ter, as highlighted in the figure below. Two of these half-strands are traversed from ori to ter in the 5' â†’ 3' direction and are thus called forward half-strands (represented by thin blue and green lines in the figure below). The other two half-strands are traversed from ori to ter in the 3' â†’ 5' direction and are thus called reverse half-strands (represented by thick blue and green lines in the figure below).

##  1.2 Asymmetry of Replication

DNA replication can be tricky to understand if you're new to biology. But donâ€™t worry  the main idea you need to remember is this:

Replication is not the same in both directions. The two halves of the DNA strand (called the forward and reverse half-strands) are copied in different ways.

Hereâ€™s why:

- DNA is made up of two strands, and each strand has a direction â€” 5' to 3' or 3' to 5'.

- The enzyme that copies DNA, called DNA polymerase, can only move in the 3' â†’ 5' direction along the strand it is reading.

This means:

- On the reverse half-strand, the polymerase can move smoothly from the starting point (ori) to the endpoint (ter) without any problem.

- But on the forward half-strand, the polymerase canâ€™t go in the same direction. Instead, it has to work in short chunks, going backward toward the ori.

The figure below shows how this works visually  and why replication happens differently on each half.



**Figure:** Replication begins at ori (primers shown in red) with the synthesis of fragments on the reverse half-strands (shown by thick lines). A DNA polymerase must wait until the replication fork has opened some (small) distance before it starts copying the forward half-strands (shown by thin lines) backward to ori.


On a forward half-strand, DNA replication works in short steps:

- The replication fork (where the DNA is being unzipped) must open up a bit  about 2,000 nucleotides.

- Once there's enough space, a primer is added near the newly opened section.

- Then, the DNA polymerase starts copying a small segment of DNA, moving backward toward the origin (ori).

This process keeps repeating: open a little more, add a primer, copy backward.
By the time both DNA polymerases reach the origin from opposite directions, you get the situation shown in the figure below.



**Figure:** The daughter fragments are now synthesized (with some delay) on the forward half-strands (shown by thin lines).




<span style="background-color: yellow; color: black;">**Note the contrast between the correct replication figure (top) and the incorrect replication figure (bottom).**</span>

After this stage, replication continues smoothly on reverse half-strands, since DNA polymerase can move along without stopping.

But on forward half-strands, the polymerase has to keep waiting for the fork to open another ~2,000 nucleotides. Each time this happens:

- A new primer is added.

- DNA polymerase copies another short fragment back toward the origin (ori).

This stop-and-start process creates short pieces of DNA, known as Okazaki fragments, which are built in chunks rather than one long strand.
You can see these fragments forming in the figure below.



**Figure:** The replication fork continues growing. Only one primer is needed for each of the reverse half-strands (shown by thick lines), while the forward half-strands (shown by thin lines) require multiple primers in order to synthesize Okazaki fragments. Two of these primers are shown in red on each forward half-strand.

When the replication fork reaches ter, the replication process is almost complete, but gaps still remain between the disconnected Okazaki fragments, as shown in the figure below.



Finally, consecutive Okazaki fragments must be sewn together by an enzyme called DNA ligase, resulting in two intact daughter chromosomes, each consisting of one parent strand and one newly synthesized daughter strand, as shown in the figure below. In reality, DNA ligase does not wait until after all the Okazaki fragments have been replicated to start sewing them together.



Biologists call reverse half-strands (thick lines) leading strands because DNA polymerase copies them smoothly in one go.
They call forward half-strands (thin lines) lagging strands because theyâ€™re copied in short chunks, starting and stopping often.

The terms "leading" and "lagging" can be confusing, since different books use them in different ways. Thatâ€™s why we use "reverse" and "forward" to make things simpler.

##  1.3 Peculiar Statistics of the Forward and Reverse Half-Strands

**Deamination**

In the last section, we learned that DNA is copied faster on the reverse half-strand, but more slowly on the forward half-strand. Now weâ€™ll use this difference to help us find where DNA replication starts (called ori).

But how can this difference help us?

Well, the reverse half-strand is copied quickly, so it stays double-stranded most of the time.
But the forward half-strand waits longer, so it stays single-stranded for a while.

Single-stranded DNA gets more mutations than double-stranded DNA.
So if one type of nucleotide (A, T, C, or G) mutates more easily in single strands, weâ€™ll see less of it on the forward half-strand.

This difference can help us guess where the ori is.

Following up on this thought, letâ€™s compare the nucleotide counts of the reverse and forward half-strands. If these counts differ substantially, then we will design an algorithm that attempts to track down these differences in genomes for which ori is unknown. The nucleotide counts for Thermotoga petrophila are shown in the figure below.

The amounts of A and T are almost the same on both half-strands.

But thereâ€™s a difference with C and G:

- C appears more often on the reverse half-strand:

- G appears less often on the reverse half-strand:
201,634 (reverse) âˆ’ 211,607 (forward) = âˆ’9,973

This shows a clear difference in how C and G are distributed between the two half-strands.

These differences happen because cytosine (C) can change into thymine (T) through a process called deamination.

This happens much more often when DNA is single-stranded â€” about 100 times faster.

As a result:

- There is less cytosine (C) on the forward half-strand.

- This also causes guanine (G) to drop on the reverse half-strand, because the forward strand creates the reverse strand during replication.

So overall, this explains why we see fewer Cs and Gs where DNA is single-stranded for longer.

Letâ€™s try using the unusual pattern caused by deamination to find the origin of replication (ori) in a circular bacterial genome.

We know that cytosine (C) is found more often in one half of the genome and less often in the other half. So weâ€™ll slide a large window across the genome, with size equal to half of the genomeâ€™s length (len(Genome)//2 in Python), and count how many Cs are in each window.

In Python means integer division. For example, 11//2 gives 5 (not 5.5).

Our goal:

- The window with the fewest Cs should line up with the forward half-strand,

- The one with the most Cs with the reverse half-strand.

If we can spot these two halves, we can guess where ori is!

**Analyzing a genomeâ€™s half-strands**

Most bacteria have circular genomes, but we were treating them like straight lines before  and that was okay because our window was small.

Now that we're using a huge window (half the genome), we need to handle the fact that the genome loops around.

To do this, weâ€™ll make a new string called ExtendedGenome.
Itâ€™s just the genome plus the first half of it again at the end.

Example:
If the genome is:

<p align="center"><code>"CTGCTTCGCCCGCCGGACCGGCCTCGTGATGGGGT"<code></p>


Then ExtendedGenome would be:


<p align="center"><code>"CTGCTTCGCCCGCCGGACCGGCCTCGTGATGGGGTCTGCTTCGCCCGCCGGA"<code></p>





We want to count how many Câ€™s (cytosines) are in each big window of the genome.

To do this, we will use a list called symbol_array.

- Each number in this list tells us:
"How many C's are in the window starting at this spot?"

- The window size is half the genome length (len(Genome)//2).

So if we slide a window across the genome (starting from position 0, 1, 2, and so on), symbol_array[i] will tell us how many C's are in the window starting at position i.

Thatâ€™s it  weâ€™re just counting C's in every window!

Fortunately, we already know how to count the number of occurrences of C in a window of ExtendedGenome: use the function PatternCount()! We can therefore define the following function that takes strings Genome and symbol as input and returns the symbol array of Genome corresponding to symbol

```python
#recreating the pattern count function that we created in previous notebook
def PatternCount(Pattern, Text):
    """ 
    input: it will takes two inputs, Pattern(symbol or one pattern), Text(Dna sequence or text)
    Output: it will output us the count of occurance of each symbol in text
    """
    #start counter with zero
    count = 0
    n = len(Text)
    k = len(Pattern)
    for i in range(n-k+1):
        if Text[i:i+k]  == Pattern:
            count += 1
    return count

#defining the symbol arraye function
def SymbolArray(Genome, symbol):
    """ 
    input: it will takes two input , Dna sequnece and pattern or symbol
    output: it will return us dictionary or array and with the count of occurance
    """
    #initialize empty dictionary to store things later
    array = {}
    n = len(Genome)
    #make extended genome to assigne it to vairable
    ExtendedGenome = Genome + Genome[0:n//2]
    for i in range(n):
        array[i] = PatternCount(symbol,ExtendedGenome[i:i+(n//2)])
    return array  



#lets test these functions 
Genome = "AAAAGGGG"
Symbol = "A"

SymbolArray(Genome, Symbol)
```

The algorithm checks each part of the genome multiple times. For example:

- First, it looks at the first n//2 positions of the genome.

- Then, it moves one step forward and checks the next n//2 positions, and repeats this process.

This repeated checking makes the algorithm slow.

Can we speed it up? Yes! Instead of going back and forth, we can find a way to look at the genome just once. This would only take a smaller number of comparisons and make the process much faster, reducing the running time from days to seconds.

To make SymbolArray() faster, we notice that when we move the window one symbol to the right, the number of times the symbol appears in the window doesn't change much. So, instead of starting over each time, we can update the count efficiently.

For example, let's say we have the string "CTGCTTCGCCCGCCGGACCGGCCTCGTGATGGGGTATGCTTCGCCCGCCGGA". The window of symbols starting at position 0 is "CTGCTTCGCCCGCCGG". Now, if we shift the window to position 1, it becomes "TGCTTCGCCCGCCGGA". Instead of counting all occurrences of the symbol from scratch, we can just:

- Remove the symbol that is no longer in the window (for example, C at position 0),

- Add the symbol that just entered the window (for example, A at the new end of the window).

This way, when we shift the window, we only need to update the count by subtracting 1 (for the removed symbol) and adding 0 (for the new symbol). If we already know the count for position 0 is 8, we automatically know that the count for position 1 is 7. This makes the process much faster!

```python
def FasterSymbolArray(Genome, symbol):
    """
    input: Genome Sequence and symbol that we are looking for
    output: array of occurances of that symbol
    """
    #initialize empty dict
    array = {}
    n = len(Genome)
    ExtendedGenome = Genome + Genome[0:n//2]
    #look at the first hald of Genome to compute first array value
    array[0] = PatternCount(symbol, Genome[0:n//2]) #using pattern count function
    for i in range(1, n):
        #start by setting the current array value equal to the previous array value
        array[i] = array[i-1]
        #the current array value can differ from the previous array value by at most 1
        if ExtendedGenome[i-1] == symbol:
            array[i] = array[i] - 1
        if ExtendedGenome[i+(n//2)-1] == symbol:
            array[i] = array[i] + 1
    return array

Symbol = "C"
result = FasterSymbolArray(Genome, Symbol)
```

##  1.4 The Skew Diagram

In the table containing nucleotide counts for T. petrophila (reproduced below), we noted that not just C but also G has peculiar statistics on the forward and reverse half-strands.



In practice, scientists use a more accurate approach that accounts for both G and C when searching for ori. As the above figure illustrates, the difference between the total amount of guanine and the total amount of cytosine is negative on the reverse half-strand and positive on the forward half-strand.

Thus, our idea is to traverse the genome, keeping a running total of the difference between the counts of G and C. If this difference starts increasing, then we guess that we are on the forward half-strand; on the other hand, if this difference starts decreasing, then we guess that we are on the reverse half-strand (see figure below).

We want to keep track of how many G's and C's we've seen in the genome so far. To do this, we make a list called the skew array.

- We start with Skew[0] = 0.

- Then we go through each letter in the genome one by one.

  - If we see A or T, the skew stays the same.

  - If we see G, we add 1 to the skew.

  - If we see C, we subtract 1 from the skew.

So at each step, the skew array tells us the difference between the number of G's and C's up to that point in the genome.

```python
#lets write a function for this
def SkewArray(Genome):
    """
    input: genome sequence
    output: the skew array of the Genome as a list
    """
    #starting skew at zero position
    skew = [0]
    n = len(Genome)
    for i in range(n):
        if Genome[i] == "G":
            skew.append(skew[-1] + 1)
        elif Genome[i] == "C":
            skew.append(skew[-1] - 1)
        else:
            skew.append(skew[-1])
    return skew


#lets test the function
Genome = "CATGGGCATCGGCCATACGCC"
SkewArray(Genome)
```

```python
# Get skew values
skew = SkewArray(Genome)

# Create position list
positions = list(range(len(skew)))

# Plot
plt.figure(figsize=(10, 4))
plt.plot(positions, skew, linewidth=1, color='blue')
plt.xlabel('Position')
plt.ylabel('Skew Value')
plt.title('Skew Array Plot')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Figure:** (Top) The skew array for Genome equal to "CATGGGCATCGGCCATACGCC", reproduced from the previous step. (Bottom) The skew diagram corresponding to Genome. The skew increases when we encounter G and decreases when we encounter C.

<span style="background-color: orange; color: black;">**After looking at the E. coli skew diagram, where do you think that ori is located in E. coli?**</span>

The ori (origin of replication) in the E. coli genome is typically located at the point where the skew diagram transitions from negative to positive. This transition occurs at the minimum of the skew diagram, as the graph reflects the difference between the counts of G and C nucleotides, which changes from negative (on the reverse half-strand) to positive (on the forward half-strand).

From the figure, the skew diagram for E. coli reaches its minimum around position 3923620 and its maximum around position 1550413. Based on the observed pattern, ori is most likely located at the point where the skew transitions from negative to positive, which in this case is around position 3923620.

This is because the transition marks the shift from the reverse half-strand (which has more C than G, reflected by negative skew) to the forward half-strand (which has more G than C, reflected by positive skew).

Letâ€™s follow the 5' â†’ 3' direction of DNA and walk along the chromosome from ter to ori (along a reverse half-strand), then continue on from ori to ter (along a forward half-strand). In the figure below, we see that the skew is decreasing along the reverse half-strand and increasing along the forward half-strand. Thus, the skew should achieve a minimum at the position where the reverse half-strand ends and the forward half-strand begins, which is exactly the location of ori!




We have just developed an insight for a new algorithm for locating ori: it should be found where the skew array attains a minimum.

**Minimum Skew Problem:** Find a position in a genome where the skew diagram attains a minimum.

- **Input:** A DNA string Genome.
- **Output:** All integer(s) i minimizing Skew[i] among all values of i (from 0 to len(Genome)).

We can outline the following function to find these minimum skew positions, using our subroutine SkewArray(Genome) that generates the skew array from a genome.

```python
#lets write the function
def MinimumSkew(Genome):
    """ 
    input : genome sequence
    output: output the list of position where there is minimum skew
    """
    #initialize the empty list
    positions = []
    #use the function of skew array we created above and assign it to variable
    sarray = SkewArray(Genome)
    minarr = min(sarray)
    for i in range(len(sarray)):
        if sarray[i] == minarr:
            positions.append(i)
    return positions


#let test the function
Genome = "TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT"
min_skew = MinimumSkew(Genome)
print(min_skew)
```

**Hamming Distance Problem:** Compute the Hamming distance between two strings.

**Input:** Two strings of equal length.

**Output:** The Hamming distance between these strings

```python
def HammingDistance(p, q):
    """ 
    input: two strings
    output: return us the count of hamming distance
    """
    count = 0
    #iterate over the len of p string
    for i in range(len(p)):
        #if a character of string p is not as same as character of p
        if p[i] != q[i]:
            #count and increment on every step
            count += 1
    return count


#lets check the functions
p = "GGGCCGTTGGT"
q = "GGACCGTTGAC"

HammingDistance(p, q)
```

A k-mer (short DNA string) Pattern is said to appear in a DNA sequence Text with up to d mismatches if there's a part of Text (same length as Pattern) that differs from Pattern in d or fewer places. In other words, the Hamming distance between them is at most d. Since DnaA boxes can have small changes, we need to look for patterns that match closely, not just exactly.

**Approximate Pattern Matching Problem:** Find all approximate occurrences of a pattern in a string.

**Input:** Strings Pattern and Text along with an integer d.

**Output:** All starting positions where Pattern appears as a substring of Text with at most d mismatches.

```python
def ApproximatePatternMatching(Text, Pattern, d):
    """ 
    input: we have 3 inputs. Text means Dna sequance, Pattern means a given pattern, and d means a value for maximum or minimum hamming distance value.
    output: return us the list of containing all positions where we found pattern with defined hamming distance d
    """
    #initialize empty list of positions
    positions = []
    n = len(Text)
    k = len(Pattern)
    #range over the range of n-k+1
    for i in range(n-k+1):
        substring = Text[i:i+k]
        #find the hamming distance between Given Pattern and subtring that we get
        if HammingDistance(Pattern, substring) <= d:
            #if the distance is less than or equal to given values of d then append to the list
            positions.append(i)
    return positions


#lets check the functions
Text = "CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT"
Pattern = "ATTCTGGA"
d = 3 #means minimum 3 words difference would be acceptable

ApproximatePatternMatching(Text, Pattern, d)
```

Our goal now is to modify our previous algorithm for the Frequent Words Problem in order to find DnaA boxes by identifying frequent k-mers, possibly with mismatches. Given input strings Text and Pattern as well as an integer d, we extend the definition of PatternCount() to the function ApproximatePatternCount(Pattern, Text, d). This function computes the number of occurrences of Pattern in Text with at most d mismatches.

For example,

**ApproximatePatternCount(AAAAA, AACAAGCATAAACATTAAAGAG, 1) = 4**

because "AAAAA" appears four times in this string with at most one mismatch: "AACAA", "ATAAA", "AAACA", and "AAAGA". Notice that two of these occurrences overlap.

```python
#in this function we slightly need a modification
def ApproximatePatternCount(Text, Pattern, d):
    """ 
    input: three inputs as before Text(dna sequance), Pattern(a defined pattern), d(an acceptable number of different element)
    output: we will get the count of the the occurance of element
    """
    count = 0
    n = len(Text)
    k = len(Pattern)
    for i in range(n-k+1):
        substring = Text[i:i+k]
        if HammingDistance(Pattern, substring) <= d:
            count += 1
    return count


#lets check the functions
Text = "CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT"
Pattern = "ATTCTGGA"
d = 3 #means minimum 3 words difference would be acceptable

ApproximatePatternCount(Text, Pattern, d)
```
