## 1.1 A Journey Of Thousand Miles


DNA replication is a crucial process in a cell. Before a cell divides, it needs to make a copy of its DNA so each new cell gets its own DNA. In 1953, Watson and Crick discovered the structure of DNA, the famous double helix. They suggested that during replication, the two DNA strands separate, and each acts as a template to make a new strand. This results in two pairs of matching DNA strands.

For example, in DNA, the bases pair up in a specific way: Adenine (A) pairs with Thymine (T), and Cytosine (C) pairs with Guanine (G). So, if one strand has a sequence like A T G C, the other strand will have T A C G as its complementary sequence.

DNA replication might seem simple, but it's actually a very complicated process. Watson and Crick first described the basic idea, but there’s a lot more going on behind the scenes to make sure the DNA is copied correctly.

At first, a computer scientist might think DNA replication is like copying a string of text—just take the genome and make a copy. However, by looking closer at how DNA replication works in real life, we can discover new ways to improve algorithms for analyzing replication and other biological processes.

Replication starts at a specific spot in the genome called the replication origin (ori). This is where the process of copying DNA begins, and it's carried out by special proteins called DNA polymerases. Understanding where ori is located is crucial for many biological studies and medical treatments.

For example, gene therapy often uses viral vectors—engineered mini-genomes that can enter cells, much like real viruses. These vectors can carry new genes into the cells, which can then be used to treat diseases. In gene therapy, the vector needs to replicate inside the cell to produce enough therapeutic proteins to help treat the disease. To make sure the vector replicates properly, scientists need to know exactly where the ori is in the vector's genome and make sure nothing interferes with it.

<span style="background-color: red; color: black;">**Problem: Finding the Starting Point of DNA Replication**</span>



In a DNA string (which is a sequence made up of the letters {A, C, G, T}), the replication process starts at a specific point called the origin of replication (ori). The task is to find where this starting point (ori) is located within the DNA string.

**Here's the problem broken down:**

**Input:** A DNA string (just a long sequence of letters A, C, G, and T).

**Output:** The position (or location) of ori (where the replication begins).

**Example:**

If we have a DNA string, we need to find where ori starts in it. This is similar to searching for a specific word in a sentence.

<span style="background-color: yellow; color: black;">**Why is this a clear computational problem?**</span>



This problem is clear because:

We have an input (the DNA string).

We need to output the position where ori starts in that string.

To solve this, we need to find where the replication starts in the sequence, which is like finding a pattern in a string.

Even though the Finding the Origin of Replication problem is an important biology question, it’s not clear enough for computer scientists to solve right away. Biologists might try to find the starting point of DNA copying (called ori) by deleting small parts of DNA and seeing what happens. But computer scientists would want more details before starting.

So why do biologists need computer help? Because computers can solve problems much faster than experiments in the lab. Also, many lab results only make sense with computer analysis. Finding ori through lab work takes a lot of time and effort. That’s why we want to use smart computer methods to find ori  so scientists can save time and focus on other important work.

###  Counting a Pattern in DNA

Before finding frequent words in DNA, we first need to count how many times a **small pattern** (like `"ATG"`) appears in a **longer DNA string**.

We’ll write a simple program called `PatternCount(Pattern, Text)` to do this.

1. Start with a count of `0`:  
<p align="center"><code> count = 0 </code></p>

2. Move a window across the whole text, one letter at a time.
3. Each time the window text matches the pattern, increase the count by 1.



**Figure:** Sliding a window to compute PatternCount(Pattern, Text), which is equal to 3, for Pattern equal to "ATA" and Text equal to "CGATATATCCATAG". We initialize the value of count to be equal to zero and then increment it each time that Pattern appears in Text (shown in green).

Before we start sliding the window over the entire string, let's first focus on the simpler problem of checking if a smaller pattern (Pattern) matches a specific part (or k-mer) of the text (Text).

In Python, we can get a part of the string (called a k-mer) from Text using this syntax:
<p align="center"><code> Text[i:i+k] </code></p>

Where i is the starting position, and k is the length of the pattern. For example:

- If Text is "GACCATACTG" and i = 4 and k = 3, then Text[4:7] will give us "ATA".

Remember:

- Python starts counting from 0, so the first letter of Text is at position 0, not 1.

- The last letter of Text is at position len(Text)-1.

This will help us check if the pattern fits into the string.

## 1.3 Hidden Messages in the Replication Origin (part 2)

<span style="background-color: yellow; color: black;">**The Frequent Words Problem**</span>


A most frequent k-mer is the DNA word of length k that appears the most in a given string.

Example:

- Most frequent 5-mer in "ACAACTATGCATACTATCGGGAACTATCCT" → "ACTAT"

- Most frequent 3-mer in "CGATATATCCATAG" → "ATA"



<span style="background-color: orange; color: black;">**Find the most frequent 2-mer of "GATCCAGATCCCCATAC". (Lets try to solve by hand not by python)**</span>



To find the most frequent 2-mer (a pattern of length 2) in the given string "GATCCAGATCCCCATAC", we need to follow these steps:
- Step 1: Generate all possible 2-mers from the string "GATCCAGATCCCCATAC". The 2-mers are: GA, AT, TC, CC, CA, AG, AT, CC, CC, CC, CA, TA, AC

- Step 2: Count the occurrences of each 2-mer. GA: 1 AT: 2 TC: 2 CC: 4 CA: 2 AG: 1 CC: 4 (combined with the previous count) AC: 1

- Step 3: Identify the 2-mer(s) with the highest count. The most frequent 2-mer is CC, occurring 4 times.

Therefore, the most frequent 2-mer of "GATCCAGATCCCCATAC" is CC.

We now have a clear computer problem: find the most frequent words in DNA.

A computational problem means:

- What we give the computer (input)

- What we want it to return (output)

This helps us tell the computer exactly what to do

**Frequent Words Problem:** Find the most frequent k-mers in a string.

 **Input:** A string Text and an integer k.
 
 **Output:** All most frequent k-mers in Text.

To find the most frequent k-mers in a string `Text`, we count how many times each k-mer of length `k` appears.

For example, in the string `"CGATATATCCATAG"` with `k = 3`, we extract all 3-letter substrings (k-mers) and count their occurrences.

We store these counts in a **frequency map** (a dictionary where:
- keys = k-mers
- values = number of times that k-mer appears in the text)

Once we build this map, we can find the most frequent k-mer by identifying the one with the highest count.


<p align="center"><code>

GAT → 1<br>

</code></p>


Fortunately, Python allows us to easily implement a frequency map by providing a built-in data structure called a dictionary.  You can think of a dictionary as a set of keys (first column in the figure above), where each key refers to a value (second column in the figure above). 

We are now ready to write a function FrequencyMap() that generates a frequency map from a given string Text and integer k.  To do so, we will first generate a map that forms a key for every k-mer appearing in Text and assign it a value of 0, as illustrated below.

<p align="center"><code>

So let us first generate an empty dictionary.
<p align="center"><code> freq = {} </code></p>

Next, we will range through the text and assign every pattern a value of 0.  We already know how to range through every k-mer in a string from our work with  PatternCount() in a previous module.  

<left>

```python

(Note that this code may assign the same key more than once, which is fine.)  The code below is therefore the start of our FrequencyMap() function.

<left>

```python
def FrequencyMap(Text, k):
    freq = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern] = 0
    return freq

Now that we can generate the beginning of a frequency map, it remains to range a sliding window of length k through the text again. For each such k-mer Pattern, we need to increase the current value of freq[Pattern] by 1.

````python
def FrequencyMap(Text, k):
    """
    Input: Text(string) and k (length of k-mer(which is 3 in this case))
    Output: return us dictionary with key of k-mer and value of frequncy(number of occurance)
    """
    #lets first instantiate the empty dictionary after that we will append
    freq = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        if Pattern in freq:
            freq[Pattern] += 1
        else:
            freq[Pattern] = 1   
    return freq


#lets utilize this function 
#initiate variable (we can directly put the values, but its better to initiate variable first)
k = 3
Text = "CGATATATCCATAG"
print((FrequencyMap(Text, k)))
````

````python
#importing matplot lib liabrary to visualize the data
import matplotlib.pyplot as plt

#initialize the function the we creatd above and assigne it to variable for visualization purpose
freq_map = FrequencyMap(Text, k)

plt.figure(figsize=(10, 5))
bars = plt.bar(freq_map.keys(), freq_map.values(), color='skyblue', edgecolor='black')
plt.xlabel('k-mers')
plt.ylabel('Frequency')
plt.title(f'Frequency of {k}-mers in Text')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Highlight the max k-mer(s)
max_freq = max(freq_map.values())
for bar, label in zip(bars, freq_map.keys()):
    if freq_map[label] == max_freq:
        bar.set_color('orange')

plt.tight_layout()
plt.show()
````

Now that we have a function to generate a frequency map from a given string and integer, we need to find the maximum value in this map and return all keys that achieve the maximum; this will represent a function FrequentWords(Text, k) to solve the Frequent Words Problem.

For example, given Text equal to "CGATATATCGAT" and k equal to 3, we have the following frequency map, and we would like to return a list containing all keys of this dictionary achieving the maximum value of 2

<p align="center"><code>
ATA → 2<br>
ATC → 1<br>
CGA → 1<br>
GAT → 2<br>
TAT → 2
</code></p>

So let us start by generating an empty list, which will eventually hold the most frequent k-mers.

<p align="center"><code> words = [] </code></p>

To generate the frequency map for Text and k, we use the FrequencyMap() function we created earlier. This can be used inside another function as a subroutine.

<p align="center"><code> freq = FrequencyMap(Text, k) </code></p>

To find the most frequent k-mers in Text, we need to find the highest value in the Count dictionary. Python has a built-in function values() that returns all the values in a dictionary, and max() to find the highest value in a list. So, we can use these to find the maximum value in the freq map like this:

<p align="center"><code> m = max(freq.values()) </code></p>

Finally, it remains to find every key in the frequency map whose corresponding value is equal to the maximum m.  To do so, we will need to range over all keys in freq

<p align="center"><code> for key in freq: </code></p>

````python
def FrequentWords(Text, k):
    """
    input: Text and kmer
    output:  most frequent kmer 
    """
    #initialize empty list
    words = []
    #use FrequencyMap function the we created above and assign it to variable
    freq = FrequencyMap(Text, k)
    #getting max value and assign it to variable
    m = max(freq.values()) 
    #iterate over the freq variable
    for key in freq:
        if freq[key] == m:
            words.append(key)
    return words


#lets use this function
# Now, we will set Text equal to the oriC of Vibrio cholerae and Pattern equal to "TGATCA"
Text = "ATCAATGATCAACGTAAGCTTCTAAGCATGATCAAGGTGCTCACACAGTTTATCCACAACCTGAGTGGATGACATCAAGATAGGTCGTTGTATCTCCTTCCTCTCGTACTCTCATGACCACGGAAAGATGATCAAGAGAGGATGATTTCTTGGCCATATCGCAATGAATACTTGTGACTTGTGCTTCCAATTGACATCTTCAGCGCCATATTGCGCTGGCCAAGGTGACGGAGCGGGATTACGAAAGCATGATCATGGCTGTTGTTCTGTTTATCTTGTTTTGACTGAGACTTGTTAGGATAGACGGTTTTTCATCACTGACTAGCCAAAGCCTTACTCTGCCTGACATCGACCGTAAATTGATAATGAATTTACATGCTTCCGCGACGATTTACCTCTTGATCATCGATCCGATTGAAGATCTTCAATTGTTAATTCTCTTGCCTCGACTCATAGCCATGATGAGCTCTTGATCATGTTTCCTTAACCCTCTATTTTTTACGGAAGAATGATCAAGCTGCTGCTCTTGATCATCGTTTC"
k = 10 #length of the k-mer

#print that function 
print(FrequentWords(Text, k))
````

The reverse complement of a DNA string Pattern is the string formed by taking the complementary nucleotide of each nucleotide in Pattern, then reversing the resulting string. ````python
def Reverse(Pattern):
    """ 
    input: Pattern of string 
    output: reverse the string
    """
    #initialize the empty string z
    z = ""
    #iterate over the pattern
    for char in Pattern:
        #append into z with char of Pattern
        z = char + z
    return z
````

Pattern Matching Problem: Find all occurrences of a pattern in a string.

 **Input:** Strings Pattern and Genome.

 **Output:** All starting positions in Genome where Pattern appears as a substring.

````python
def PatternMatching(Pattern, Genome):
    """ 
    Input: Pattern like ("ATAT"), and Genome like String of Dna 
    Output: This positions will output the list of positions of the the Pattern to be found in string (start a zero)
    """
    #this function is as as counting problem
    #initialize the empty list of positions
    Positions = []
    n = len(Genome)
    k = len(Pattern)
    for i in range(n-k+1):
        if Genome[i:i+k] == Pattern:
            Positions.append(i)
    return Positions


#lets test the functions
Pattern = "ATAT"
Genome = "GATATATGCATATACTT"
print(PatternMatching(Pattern, Genome))
````

**The Clump Finding Problem**

Imagine you’ve just sequenced a new bacterial genome and want to find the ori (origin of replication). Looking for known patterns like "ATGATCAAG" / "CTTGATCAT" or "CCTACCACC" / "GGTGGTAGG" might not work—this new genome might use a completely different hidden signal!

But don’t worry—we can try a different approach.

Instead of searching for specific patterns, we’ll look for any pattern that appears frequently in a short region of the genome. This might help us find where the ori is located.

Here’s the plan:

We slide a window of a fixed length L (like 500 bases) across the genome.

In each window, we look for any k-mer (short DNA sequence of length k) that shows up multiple times.

If a k-mer appears t or more times within a window of size L, we call it an (L, t)-clump.

Example:
If we set:

k = 4 (looking for 4-letter patterns)

L = 25 (window size of 25 bases)

t = 3 (must appear at least 3 times)

Then the pattern "TGCA" is considered a (25, 3)-clump in this DNA string:

<p align="center"><code>
"gatcagcataagggtccCTGCAATGCATGACAAGCCTGCAGTtgttttac"<code></p>

This method can help uncover hidden repeated patterns that might mark the origin of replication even if we’ve never seen them before.

In our previous examples:

- The pattern "ATGATCAAG" appears 3 or more times within a 500-base region in the Vibrio cholerae genome.

- The pattern "CCTACCACC" does the same in Thermotoga petrophila.

These are both examples of (500, 3)-clumps—patterns that appear at least 3 times in a 500-base window.

**Clump Finding Problem (Simplified):**

**Goal:** Find all patterns (k-mers) that appear often in a small region of DNA.

**Input:**

- A DNA string called Genome

- A k-mer length k

- A window size L

- A minimum count t

**Output:**

All different k-mers that appear at least t times within any window of length L in the genome.

These are the k-mers that form (L, t)-clumps.

## 1.6 Detour: The Most Beautiful Experiment in Biology

The Meselson-Stahl experiment, done in 1958 by Matthew Meselson and Franklin Stahl, is often called “the most beautiful experiment in biology.”

Back then, scientists weren’t sure how DNA copies itself. They had three main ideas:

1. **Semiconservative model:** Each original DNA strand acts as a guide to build a new strand. So, every new DNA molecule has one old and one new strand.

2. **Conservative model:** The entire original DNA stays together, and a completely new DNA copy is made from it. This means one DNA has only old strands, and the other has only new strands.

3. **Dispersive model:** The original DNA breaks into pieces, mixes with new DNA, and the result is a blend of old and new bits throughout both strands.

Meselson and Stahl designed their experiment to find out which of these three ideas was correct.

Meselson and Stahl had a smart idea: they used two types of nitrogen—Nitrogen-14 (¹⁴N), which is lighter and more common, and Nitrogen-15 (¹⁵N), which is heavier. Since DNA naturally contains nitrogen, they used this difference to track DNA replication.

First, they grew E. coli bacteria in a ¹⁵N medium (a food source with heavy nitrogen). Over time, the bacteria took in the heavy nitrogen, making their DNA heavier.

Once they were sure the DNA was fully heavy, they moved the bacteria into a normal ¹⁴N medium (with lighter nitrogen).

 <span style="background-color: orange; color: black;">**What do you think happened when the “heavy” E. coli replicated in the “light” 14N medium?**</span>

We learned earlier that the complement strand is constructed from the free flowing nucleotides - and the atoms for those free necleotides will obviously have to come from the external environment. So when the heavy N-15 E. coli split in the light N-14 environment, there are three things that could have happened:

**1.** If the conservative way was true - then one daughter cell would inherit both heavy N-15 strands and the other daughter would copy those strands using the lighter N-14. The outcome? One daughter cell would be heavier than the other.

**2.** If the semi-conservative way was true - then both daughters would inherit one N-15 strand and construct the missing complement using N-14. Both then have one N-15 strand and one N-14 strand. The outcome? Both daughters weigh the same.

**3.** If the dispersive way was true - then both daughters an inherit an equal number of pieces of N-15 strands and fill in the gaps with N-14. Both then have a patchwork of N-15 and N-14. The outcome? Again both weigh the same.

Therefore, if a difference in weight is observed then 2 and 3 are refuted. 1 is true. If no weight difference is observed, the weights are the same, 1 is refuted. Either 2 or 3 is true.

Of course we already know that the semi-conservative view is how it works so no weight difference will be observed after the split

The genius of the Meselson-Stahl experiment was that any new DNA made after the switch to ¹⁴N would only contain the lighter nitrogen. The three DNA replication models predicted different results based on this.

**After one round of replication:**

The conservative model said half the DNA should stay heavy (with ¹⁵N) and the other half should be light (with ¹⁴N), so there should be two different DNA weights.

But when they spun the DNA in a centrifuge, all the DNA had the same weight—right in between heavy and light. This result disproved the conservative model.

However, they couldn’t yet rule out the other two models—semiconservative and dispersive—because both predicted this "mixed" weight after one round.

 <span style="background-color: orange; color: black;">**What would the dispersive and semiconservative models predict about the density of E. coli DNA after two rounds of replication?**</span>

Let's see. So we have crossed out one possibility (the conservative view) by virtue of the daughter DNA's weighing the same. Now two sets of daughter cells could be present. One set (the semi-conservative set) where one strand is wholly N-15 strand and another is wholly N-14. The other set (the dispersive set) where both strands are a patchwork of N-15 and N-14. Upon splitting:

1. The semi-conservative set will have one daughter with one strand wholly N-15 and the complement wholly N-14. The other daughter will have both her strands wholly N-14. So one daughter is heavier than the other.

2. The dispersive set will have both daughters having one strand a patchwork and the other wholly N-14. So both daughters weigh the same.

Therefore, if a weight difference is observed then semi-conservative view is correct and dispersive is refuted. If no weight difference is observed then the semi view is refuted and the dispersive view is correct.

Let’s look at the dispersive model. It says each new DNA strand is a mix of old and new pieces—like a mash-up. So, after two rounds of copying, each strand should be about 25% heavy (¹⁵N) and 75% light (¹⁴N). That means all the DNA should still have the same weight.

But that’s not what Meselson and Stahl saw!

After two rounds, the DNA separated into two different weights. This matched exactly what the semiconservative model predicted:

After one round: all DNA has one heavy and one light strand.

After two rounds: half the DNA is still mixed (one heavy, one light), and half is fully light (two light strands).

So the results supported the semiconservative model and disproved the dispersive one.



Meselson and Stahl had already ruled out the conservative and dispersive models of DNA replication. But they still wanted solid proof that the semiconservative model was correct. According to this model, after three rounds of replication, about 25% of the DNA should still contain a heavy (¹⁵N) strand, making it have intermediate weight, while the other 75% should be fully light (with only ¹⁴N).

That’s exactly what they saw in the lab.

This confirmed the semiconservative model of DNA replication — and it’s still accepted today.

And if anyone still has doubts, they can hear about this beautiful experiment straight from Meselson and Stahl themselves!

## 1.7 Detour: Directionality of DNA Strands

The sugar part of a nucleotide has a ring made of five carbon atoms. These atoms are labeled 1’, 2’, 3’, 4’, and 5’. The 5’ atom connects to the phosphate group in the nucleotide and eventually links to the 3’ end of the neighboring nucleotide. The 3’ atom then connects to another nucleotide in the chain. This creates two ends of the nucleotide: the 5’-end and the 3’-end (pronounced "five prime end" and "three prime end").

When looking at the entire DNA structure, each DNA strand has one end with a 3’ atom and the other end with a 5’ atom. A DNA strand is always read from the 5' → 3' direction. In the two complementary strands of the double helix, these directions run in opposite ways.
